1) O(n)

To print
n number of elements in Python and JavaScript, you can use loops or array slicing. Below are examples for both languages:

```python
def NofElements(n):
    # The loop runs from 0 to n-1, executing exactly n iterations.
    # Time Complexity: O(n) - Linear time, as the number of iterations grows linearly with n.
    for i in range(n):
        # Printing each number takes constant time for each iteration.
        # Time Complexity (per iteration): O(1)
        print(i)  # Total time for this operation is O(n) because it is inside the loop.

# Function call to print numbers from 0 to 4.
NofElements(5)  # This will print: 0 1 2 3 4


```
